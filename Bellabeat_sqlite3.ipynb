## 1b.Average Time in Bed, Minutes Asleep, and Awake Time by Day of the Week

### Merging activity table and sleepday table

```python
con = sqlite3.connect('bellabeat.db')  
join_query = """
SELECT df_sleepDay_2.SleepDay,df_sleepDay_2.TotalMinutesAsleep, df_sleepDay_2.TotalTimeInBed
FROM df_sleepDay_2 
JOIN df_activity_2
ON df_sleepDay_2.Id = df_activity_2.Id 
"""
df_joined = pd.read_sql(join_query, con)
con.close()
df_joined['SleepDay'] = pd.to_datetime(df_joined['SleepDay'])
df_joined['DayOfWeek'] = df_joined['SleepDay'].dt.day_name()
df_joined['AwakeTimeInBed'] = df_joined['TotalTimeInBed'] - df_joined['TotalMinutesAsleep']
df_grouped = df_joined.groupby('DayOfWeek').mean()[['TotalTimeInBed', 'TotalMinutesAsleep', 'AwakeTimeInBed']]
df_grouped = df_grouped.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
df_grouped.plot(kind='bar', figsize=(9, 5),color=['red', 'yellow','green'])
plt.title('Average Time in Bed, Minutes Asleep, and Awake Time by Day of the Week', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Average Minutes', fontsize=14)
plt.grid(True)
plt.legend(title='Metric', fontsize=12)
plt.show()
```
